```python
X=z.drop(['y','ID'] ,axis=1)
y=z['y']
```

```python
Q1 = X.quantile(0.25)
Q3 = X.quantile(0.75)
IQR = Q3 - Q1
```

```python
init = False

for col_name in list(X.columns):

    low  = Q1[col_name] - 1.5 * IQR[col_name]
    high = Q3[col_name] + 1.5 * IQR[col_name]

    query_string = '{} < @low or {} > @high'.format(col_name, col_name)

    outlier_arr_loop = X.query(query_string).index
    
    if not init:
        outlier_arr = outlier_arr_loop
        init = True
    
    outlier_arr = outlier_arr.union(outlier_arr_loop)
```

```python
import sklearn.model_selection as model_selection
x_train, x_test,y_train, y_test= model_selection.train_test_split(x_pca, y,test_size=0.20, random_state=105)
```

```python
xgb_params = {
 'n_trees': 500, 
 'eta': 0.0050,
 'max_depth': 8,
 'subsample': 0.95,
 'objective': 'reg:linear',
 'eval_metric': 'rmse',
 'base_score': np.mean(y_train), # base prediction = mean(target)
 'silent': 1
}
```

```python
import xgboost as xgb
dtrain=xgb.DMatrix(x_train,y_train)
dtest=xgb.DMatrix(x_test)
```

```python
xb_model=xgb.cv(xgb_params,dtrain,num_boost_round=700,verbose_eval=50)
```

```python
print(xb_model)
```

```python
from xgboost import XGBRegressor

xb=xgb.train(xgb_params,dtrain,num_boost_round=700,verbose_eval=50)
```

```python
y_predict = xb.predict(dtest)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```
